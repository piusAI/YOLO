```python
from ultralytics import YOLO
```

```python
model = YOLO('yolo11n.pt')
def predict_callback_example(predictor):
  print(f'..callback logs..\n'
        f'model predict started..\n'
        f'prediction results : {predictor.results}...\n')

model.add_callback('on_predict_start', predict_callback_example)
model.predict(source= 'https://ultralytics.com/images/bus.jpg')
```

## Training Callback

```python
def train_callback_ex(trainer):
  print(f'...Calbback Logs...\n'
        f'model training started ....\n'
        f'epochs : {trainer.epochs}\n'
        f'batch size : {trainer.epochs}\n'
        f'device : {trainer.device}\n......\n'
        f'args : {trainer.args}\n.....')

model.add_callback('on_train_start', train_callback_ex)
if __name__ =='main':
  model.train(data = 'coco8.yaml', batch=8, workers=1, epochs=1)
```

## Export Callback

```python
def export_callback_ex(exporter):
  print(f'...Callback Logs ...\n'
        f'model export started .....\n'
        f'args : {exporter.args}\n.....')

model.add_callback('on_export_start', export_callback_ex)
model.export(format='torchscript', name='JGAi')
```
